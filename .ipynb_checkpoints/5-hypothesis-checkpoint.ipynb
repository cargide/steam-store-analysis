I have some understanding of the industry, and I want to test if RPG games that are free have more than 500 positive reviews.  

The null hypothesis in this case would be what was said above.  
The alternative hypothesis is that not being true.
My significance level is the usual 0.05, which means I am 95% confident on my hypothesis for any sample.

## 1) Selecting data


```python
free_rpgs = steam_data[((steam_data['genre1']=='RPG')|(steam_data['genre2']=='RPG')|(steam_data['genre3']=='RPG'))&steam_data['is_free']]['positive']
free_rpgs
```

## 2) Testing the hypothesis

```python
stat, pval = ttest_1samp(free_rpgs, 500)
print('stat is  ', stat)
print('pvalue for the two-tailed test is ', pval)
```

In this case, the two tailed test indicates that I need to reject my null hypothesis.
